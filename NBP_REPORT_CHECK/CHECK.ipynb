```python
EXCEL_READ_AR2 = [
    'AR2',
    '4a.R.L_PLiW2',
    '4a.R.W_PLiW2',
    '5a.R.LF_PLiW2',
    '5a.R.WF_PLiW2',
    '5a.R.SF',
    '6.ab.LiW',
    '9.R.L.MCC',
    '9.R.W.MCC'
]
EXCEL_READ_AR1 = [
    'ST.01',
    'ST.03',
    'ST.05',
    'ST.06',
    'ST.07',
    'p-dane'
]
```

```python
df_old = pd.read_excel(path + 'BSP_AR2_v.4.0_Q12023_20230421.xlsx', sheet_name=EXCEL_READ_AR2, header=None,
                         keep_default_na=False)
```

```python
df_new = pd.read_excel(path + 'BSP_AR2_v.4.0_Q12023_20230630.xlsx', sheet_name=EXCEL_READ_AR2, header=None,
                         keep_default_na=False)
```

```python
check = []
change_color = []

for sheet in EXCEL_READ_AR2:
    for i in range(df_old[sheet].shape[0]):
        for j in range(df_old[sheet].shape[1]):
            if str(df_old[sheet].iat[i, j]) == str(df_new[sheet].iat[i, j]):
                check.append("ok")
            else:
                check.append([f'{df_old[sheet].iat[i, j]} != {df_new[sheet].iat[i, j]}',sheet, i, j])
                change_color.append([sheet, i, j])
```

```python
for string in check:
    if string != 'ok':
        print(string)
```

```python
# Load the old and new workbooks
open_old = 'BSP_AR2_v.4.0_Q12023_20230421.xlsx'
wb_old = openpyxl.load_workbook(path + open_old)
open_new = 'BSP_AR2_v.4.0_Q12023_20230630.xlsx'
wb_new = openpyxl.load_workbook(path + open_new)

# Iterate through the list of cells
for cell_info in change_color:
    sheet_name, row_number, column_number = cell_info
    
    # Get the sheet object
    sheet = wb_new[sheet_name]
    
    # Get the cell object
    cell = sheet.cell(row=row_number + 1, column=column_number + 1)
    
    # Change the color of the cell to yellow
    yellow_fill = openpyxl.styles.PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")
    cell.fill = yellow_fill

# Save the modified workbook
wb_new.save(path + open_new)

# Iterate through the list of cells
for cell_info in change_color:
    sheet_name, row_number, column_number = cell_info
    
    # Get the sheet object
    sheet = wb_old[sheet_name]
    
    # Get the cell object
    cell = sheet.cell(row=row_number + 1, column=column_number + 1)
    
    # Change the color of the cell to yellow
    yellow_fill = openpyxl.styles.PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")
    cell.fill = yellow_fill

# Save the modified workbook
wb_old.save(path + open_old)
```
